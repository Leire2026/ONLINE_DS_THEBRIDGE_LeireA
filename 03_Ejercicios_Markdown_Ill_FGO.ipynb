### Ejercicio 3: Resaltados y HTML
Crea una línea que contenga:

1. La palabra "resaltado" con un fondo amarillo usando Markdown.
2. La palabra "color" en azul usando HTML.

Para este ejercicio tendrás que buscar un poco más en internet, venga que tú puedes.

Para usar el <mark>resaltado</mark> en Markdown tienes que utilizar `<mark>palabra<mark>`

Por otro lado, para resaltar con <font color="blue">color</font> azul usando HTML, se usa `<font color="blue">palabra</font>`



1. La palabra <mark>resaltado<mark> con un fondo amarillo usando Markdown
2. La palabra <font color="blue">palabra</font> en azul usando HTML
